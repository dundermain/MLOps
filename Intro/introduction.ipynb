Now this is a very crucial part. This is the CI/CD workflow for ML Models. We will go through each of the steps




Feature store:
1. It should be well thought from both business as well as tech POV
2. Try to design it in such a way that it is searchable and can be reused for other projects
3. It should be consistent between training and deployement environment. 

ML Metadata:
1. It tracks ans stores the experiment details like model version, data version, hyperparameters, architecture version etc
2. It will also store the results and inference details
3. It will also contrain all the stat details to compare the models

Source repository contains all the information that is required to deploy, like infrastructe details, machine requirements, how do you want to deploy it, dependencies
